```python
import pandas as pd
import numpy as np

from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler
```

```python
df = pd.read_csv("../data/cleaned_bank_churn.csv")
```

```python
df = pd.get_dummies(
    df,
    columns=['country','gender'],
    drop_first=True
)
```

```python
df.head()
```

```python
x = df.drop('churn', axis=1)
y = df['churn']
```

```python
print(x.shape)
print(y.shape)
```

```python
X_train, X_test, y_train, y_test = train_test_split(x, y, test_size=0.2, random_state=42, stratify=y)
```

```python
print(x.shape)
print(y.shape)
```

```python
print(X_train.shape)
print(X_test.shape)
print(y_train.shape)
print(y_test.shape)
```
