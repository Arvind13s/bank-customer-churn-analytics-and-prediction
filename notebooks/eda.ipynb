```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from scipy import stats
```

```python
df = pd.read_csv("../data/cleaned_bank_churn.csv")
```

```python
#Churn distribution
sns.countplot(x='churn', data=df)
plt.show()
```

```python
df['churn'].mean()*100
```

```python
#Country vs Churn
pd.crosstab(
    df['country'],
    df['churn'],
    normalize='index'
)*100
```

```python
sns.countplot(
    x='country',
    hue='churn',
    data=df
)
plt.show()
```

```python
#Gender vs Churn
sns.countplot(
    x='gender',
    hue='churn',
    data=df
)
plt.show()
```

```python
pd.crosstab(
    df['gender'],
    df['churn'],
    normalize='index'
)*100
```

```python
#Age Analysis
sns.boxplot(
    x='churn',
    y='age',
    data=df
)
plt.show()
```

```python
sns.histplot(
    data=df,
    x='age',
    hue='churn'
)
```

```python
#Credit Score Analysis
sns.boxplot(
    x='churn',
    y='credit_score',
    data=df
)
plt.show()
```

```python
#Balance Analysis
sns.boxplot(
    x='churn',
    y='balance',
    data=df
)
plt.show()
```

```python
#Active Member Analysis
sns.countplot(
    x='active_member',
    hue='churn',
    data=df
)
plt.show()
```

```python
#Product Analysis
sns.countplot(
    x='products_number',
    hue='churn',
    data=df
)
```

```python
#Correlation Heatmap
plt.figure(figsize=(12,8))

sns.heatmap(
    df.corr(numeric_only=True),
    annot=True,
    cmap='coolwarm'
)

plt.show()
```
